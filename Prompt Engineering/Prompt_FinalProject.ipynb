```python
import os
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from pinecone import Pinecone, ServerlessSpec

# Load the dataset
file_path = r'C:\Users\Client\Downloads\Symptom2Disease.csv'
data = pd.read_csv(file_path)

# Extract the symptom descriptions
symptoms = data['text'].tolist()

# Load a pre-trained model for embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate embeddings
print("Generating embeddings...")
embeddings = model.encode(symptoms, show_progress_bar=True)
embeddings_array = np.array(embeddings)

# Save embeddings to .npy file
embeddings_file_path = r'C:\Users\Client\Downloads\symptom_embeddings.npy'
np.save(embeddings_file_path, embeddings_array)
print(f"Embeddings saved to {embeddings_file_path}")

# Save labels to a .csv file
labels_file_path = r'C:\Users\Client\Downloads\labels.csv'
data['label'].to_csv(labels_file_path, index=False)
print(f"Labels saved to {labels_file_path}")

# Initialize Pinecone
pinecone_api_key = "pcsk_5GYuPi_LLpNfJewECArYBRRdFyV3DEWqsek4fSazUSVfAL8hNYRknuR5g5qzs7F2DoDRT7"  # Replace with your API key
pc = Pinecone(api_key=pinecone_api_key)

# Create or connect to the Pinecone index
index_name = "symptom-disease-index"
if index_name not in pc.list_indexes().names():
    pc.create_index(
        name=index_name,
        dimension=embeddings_array.shape[1],
        metric="cosine",
        spec=ServerlessSpec(cloud="aws", region="us-east-1")  # Adjust region if needed
    )
index = pc.Index(index_name)

# Prepare data for uploading to Pinecone
labels = pd.read_csv(labels_file_path)['label'].tolist()
ids = [f"id_{i}" for i in range(len(labels))]

# Batch upload embeddings
batch_size = 100
print("Uploading embeddings to Pinecone...")
for i in range(0, len(embeddings_array), batch_size):
    batch_ids = ids[i:i+batch_size]
    batch_embeddings = embeddings_array[i:i+batch_size].tolist()
    batch_metadata = [{"label": labels[j]} for j in range(i, i+batch_size)]
    index.upsert(vectors=list(zip(batch_ids, batch_embeddings, batch_metadata)))

print("All embeddings uploaded to Pinecone.")
```

```python
from sentence_transformers import SentenceTransformer

# Function to query Pinecone
def query_pinecone(input_text, top_k=5):
    # Generate embedding for the input text
    model = SentenceTransformer('all-MiniLM-L6-v2')
    input_embedding = model.encode(input_text).tolist()

    # Query the Pinecone index
    results = index.query(vector=input_embedding, top_k=top_k, include_metadata=True)

    # Return results as a list of labels and scores
    return [(match['metadata']['label'], match['score']) for match in results['matches']]

# Example query
input_symptom = "I have a rash and peeling skin on my arms"
results = query_pinecone(input_symptom, top_k=3)
print("Top results:", results)
```
